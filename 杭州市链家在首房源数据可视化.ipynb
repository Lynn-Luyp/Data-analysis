```python
path=r'F:\Data\lianjia\house.csv'
df=pd.read_csv(path)
```

## 三、查看数据

```python
df.info()
```

```python
df.head()
```

## 四、数据清洗

### 去除空值

```python
# 查看空值
df.isnull().sum()
```

```python
# 去除空值
df=df.dropna()
df.isnull().sum()
```

### 处理产权变量

```python
df['产权'].value_counts()
```

将未知量去除

```python
df=df[df['产权']!='未知']
```

### 处理区域变量

```python
df['区域'].value_counts()
```

```python
# 选取有代表性的区域进行地区汇总
def location(x):
    if "临安" in x: return "临安市"
    elif "上城" in x: return "上城区"
    elif "下城" in x: return "下城区"
    elif "江干" in x: return "江干区"
    elif "拱墅" in x: return "拱墅区"
    elif "西湖" in x: return "西湖区"
    elif "滨江" in x: return "滨江区"
    elif "萧山" in x: return "萧山区"
    elif "余杭" in x: return "余杭区"
    elif "富阳" in x: return "富阳区"
    elif "钱塘" in x: return "钱塘新区"
    else: return "其他"
    
df['区域']=df['区域'].apply(location)
```

```python
# 查看区域数据
df['区域'].value_counts()
```

### 处理单价变量

```python
df['单价']
```

单价应该是float类型，需要进行类型转换

```python
df['单价']=df['单价'].str[:-4].astype('float32')
```

```python
df['单价']
```

```python
# 查看单价变量
df['单价'].describe()
```

### 处理年限变量

```python
df['年限']
```

将“年限”变量划分为“起建时间”和“建筑类型”两个变量

```python
df['起建时间']=df['年限'].str.split('/').str[0]
df['建筑类型']=df['年限'].str.split('/').str[1]
```

```python
# 查看起建时间数据
df['起建时间'].value_counts()
```

```python
# 筛选掉未知年份建的书籍
df=df[df['起建时间']!='未知年建']
```

```python
# 查看建筑类型数据
df['建筑类型'].value_counts()
```

```python
# 筛选掉未知年份建的书籍
df=df[df['建筑类型']!='暂无数据']
```

```python
# 创建楼龄变量
df['楼龄']=2020-df['起建时间'].str[:4].astype('int32')
```

```python
df['楼龄']
```

```python
# 舍弃起建时间和年限数据
df=df.drop(['起建时间','年限'],axis=1)
```

### 处理户型变量

```python
# 查看数据
df['户型']
```

```python
df['室数']=df['户型'].str[0].astype('int32')
df['厅数']=df['户型'].str[2].astype('int32')
```

```python
# 去除房屋编码数据
df=df.drop(['房屋编码'],axis=1)
```

### 处理挂牌时间变量

```python
df['挂牌时间']
```

```python
df['Date']=pd.to_datetime(df['挂牌时间'],format='%Y-%M-%d')
df.set_index(df.Date,inplace=True)
```

### 处理朝向变量

```python
df['朝向']
```

```python
# 去除首段末端的空格
df['朝向']=df['朝向'].str.strip()
```

```python
# 查看字数的统计数据
df['朝向'].str.len().value_counts()
```

```python
# 利用窗户变量重新统计窗户朝向数据
def orientation(x):
    if "东南" in x:
        if x.count("南") > 1:
            if ("西南" in x) & ("南" not in x):
                return "东南"
            else:
                return "南"
        else:
            return "东南"
    elif "西南" in x:
        if x.count("南") >1:
            return "南"
        else:
            return "西南"
    elif "南" in x:
        return "南"
    elif "东" in x:
        return "东"
    elif "西" in x:
        return "西"        
    else:
        return "北"

df['窗户']=df['朝向'].apply(orientation)
```

```python
df['窗户']
```

```python
# 舍弃朝向变量
df.drop('朝向',axis=1,inplace=True)
```

### 处理楼层变量

```python
df['楼层']
```

```python
df=df[df['楼层'].str.contains('/')]
```

```python
# 引入楼层位置变量
df['楼层位置']=df['楼层'].str.split('/').str[0]
df=df[df['楼层位置'].isin(['低楼层','中楼层','高楼层'])]
```

```python
# 引入层数变量
df['层数']=df['楼层'].str.split('/').str[1]
df['层数']=df['层数'].str[1:-1].astype('int32')
```

```python
# 去除楼层变量
df=df.drop('楼层',axis=1)
```

### 处理装修情况变量

```python
def build(x):
    if '精装' in x:
        return '精装'
    elif '简装' in x:
        return '简装'
    elif '毛坯' in x:
        return '毛坯'
    else:
        return '其他'

df['装修情况']=df['装修情况'].apply(build)
```

### 处理面积变量

```python
df['面积']
```

```python
df['面积']=df['面积'].str[:-2].astype('float32')
```

所有变量均处理完毕

### 查看数据

```python
df.describe()
```

```python
# 去除0室0厅的数据
# 保留房子总价大于50万和小于3000万的数据
df=df[df['户型']!='0室0厅']
df=df[(df['总价/万元']>50)&(df['总价/万元']<3000)]
```

### 保存数据

```python
df.to_csv('F:\Data\lianjia\lianjia.csv')
```
