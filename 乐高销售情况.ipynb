# 乐高天猫旗舰店数据分析

## 一、查看数据

数据包括天猫乐高旗舰店的销售信息，由392行以及6列数据组成

## 二、导入库

```python
import matplotlib as plt
import numpy as np 
import pandas as pd
```

## 三、数据读取

```python
path=r'F:\Data\legao3225\天猫乐高旗舰店数据.csv'
df=pd.read_csv(path)
```

## 四、查看数据

```python
df.head()
```

```python
df.info()
```

## 五、数据处理

### 去除重复值

```python
# 查看重复值
df.duplicated()
```

```python
# 去除重复值
df.drop_duplicates(inplace=True)
```

```python
# 重新索引
df=df.reset_index()
```

```python
# 再次查看重复值
df.duplicated().sum()
```

### 价格标签数据处理/类型转换

```python
df.price
```

```python
df.price.astype('float')
```

存在数据为199.00-999.00的数据，用平均值代替

```python
# 查看含有'-'的数据
df.price.str.contains('-').sum()
```

```python
def price(x):
    if '-' in x:
        return (float(x.split('-')[1])-float(x.split('-')[0]))/2
    else:
        return x 

df.price=df.price.apply(lambda x: price(x))
```

```python
df.price=df.price.astype('float')
```

```python
# 查看价格数据
df.price
```

### 销售量标签数据处理/类型转换

```python
df.sales_num.value_counts()
```

将"无"替换为其余值的平均值

```python
# 求出其余值的平均值
df.loc[df.sales_num!='无','sales_num'].astype('int').mean()
```

```python
# 将‘无’替换为平均值194
df.sales_num=df.sales_num.str.replace('无','194')
```

```python
# 类型转换
df.sales_num=df.sales_num.astype('int')
```

## 六、查看数据

```python
df.head()
```

```python
df.info()
```

多了一列索引，舍弃索引列

```python
df=df.drop('index',axis=1)
```

## 七、加入销售额

```python
df['sales_volume']=df['sales_num']*df['price']
```

## 八、保存数据

```python
df.to_csv('F:\Data\legao3225\legoshop.csv')
```
